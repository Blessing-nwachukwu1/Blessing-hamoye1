**Q  Select columns â€˜Y2017â€™ and â€˜Areaâ€™, Perform a groupby operation on â€˜Areaâ€™. Which of these Areas had the highest sum in 2017?**

**Q  Select columns â€˜Y2017â€™ and â€˜Areaâ€™, Perform a groupby operation on â€˜Areaâ€™. Which of these Areas had the 7th lowest sum in 2017?**

```python
#Select Y2017 and Area columns
Selected_columns = ['Y2017','Area']
df_selected = df[Selected_columns]
```

```python
#Groupby 'Area' and 'Y2017'
Grouped_area = df_selected.groupby('Area')['Y2017'].sum()
Grouped_area
```

```python
#What is the Area with the 7th lowest sum in 2017
Seventh_lowest_area = Grouped_area.nsmallest(7).index[-1]
Seventh_lowest_area
```

**Q  Perform a groupby operation on â€˜Elementâ€™. What year has the highest sum of Stock Variation?**

```python
#Groupby on Element
Element_grouped = df.groupby('Element').sum()
Element_grouped
```

```python
# Filter the DataFrame to rows where 'Element' is 'Stock Variation'
stock_variation_df = df[df['Element'] == 'Stock Variation']
stock_variation_df
```

```python
# Calculate the sum for each year
yearly_sum = stock_variation_df.drop(columns=['Element', 'Area', 'Area Code', 'Item Code', 'Item','Unit','Element Code']).sum()
yearly_sum
```

```python
# Find the year with the highest sum of Stock Variation
year_with_highest_stock_variation = yearly_sum.idxmax()
year_with_highest_stock_variation
```
